```python
from math import *
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plot

k=16.7 
fi = pi*3/4 
```

```python
def f1(tetha, r): 
    dr=2*r/sqrt(77)
    return dr

def f2(t): 
    xt = tan(fi)*t
    return xt
```

```python
r01=167/55
r02=167/35

tetha1 = np.arange(0, 2*pi, 0.01)
r1_1 = odeint(f1, r01, tetha1)
r1_2 = odeint(f1, r02, tetha1)

t=np.arange(0, 10, 1)
tetha2=np.arctan(f2(t)/t)
r2 = np.sqrt(t * t+ f2(t) * f2(t))
```

```python
plot.polar(tetha1, r1_1, 'blue', label = 'Катер')
plot.polar(tetha2, r2, 'goldenrod', label = 'Лодка')

plot.legend()
```
